```python
from transformers import AutoModelForCausalLM, AutoTokenizer, set_seed,AutoModelForSeq2SeqLM
```

```python
import torch
torch.set_default_tensor_type(torch.cuda.FloatTensor)
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-large", use_cache=True)
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-large")
```

```python
# blogpost_title = 'Employee privacy policy' # replace this with your own blogpost title
# prompt = f'Draft an employment contract consisting of terms and conditions in English language. An employment contract is an agreement signed by the employee and employer (or labor union) regarding the rights, responsibilities and obligations of both parties during the period of employment.'
prompt = "Java developer resume Designed/developed APIs that processed"
prompt = "Java developer resume skills and education"
prompt = "Java developer resume work experience"
prompt = "Java developer resume achievements"
prompt = "Senior Java developer resume skills and education"
prompt = "Senior Java developer resume work experience"
prompt = "Senior Java developer resume achievements"
```

```python
prompt = "CONTRACT OF EMPLOYMENT MADE AND ENTERED INTO BY AND BETWEEN:"
prompt = "WHEREBY THE PARTIES AGREE AS FOLLOWS: 1. APPOINTMENT"
prompt = "DURATION 2.1 This agreement will become effective "
prompt = "THE EMPLOYEE’S DUTIES 3.1 The core of the EMPLOYEE’s duties towards the EMPLOYER is"
prompt = "WORK PLACE The EMPLOYEE will execute his / her duties at the following offices:"
```

```python
#contrastive search decoding strategy 
input_ids = tokenizer(prompt, return_tensors="pt").to(0) # max_new_tokens=600,
sample = model.generate(**input_ids,min_length =200 ,max_length=400, length_penalty=1.6,no_repeat_ngram_size=3,
                        top_k=150,top_p=0.82, penalty_alpha=0.6, temperature=0.9, repetition_penalty = 2.1,num_beams=6 , do_sample=True)
print(tokenizer.decode(sample[0], truncate_before_pattern=[r"\n\n^#", "^'''", "\n\n\n"]))
```
